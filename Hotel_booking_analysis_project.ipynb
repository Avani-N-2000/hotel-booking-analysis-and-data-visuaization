#### Chart - 6

````python
# Chart - 6 visualization code
room_df = df[['hotel', 'reserved_room_type', 'assigned_room_type']]
#Gets the counts of each unique value in the 'reserved_room_type' column and converts it to a list
room_value = list(room_df['reserved_room_type'].value_counts())
room_type = list(room_df['reserved_room_type'].unique())
#Computes the percentage of each reserved room type out of the total number of reservations. Each percentage is rounded to two decimal places
rooms = [ round((r/sum(room_value))*100,2) for r in room_value]
````

````python
# Apply Seaborn plots
sns.set(style="whitegrid")
colors = sns.color_palette("pastel")

# Combine room with percentage
type_with_room = [f'{room} - {pct}%' for room, pct in zip(room_type, rooms)]

# Plot
fig, ax = plt.subplots()
wedges, texts = ax.pie(rooms, colors=colors, startangle=140, radius=1.2,
                       wedgeprops=dict(width=0.4))
#startangle=140: Rotates the start of the pie chart to 140 degrees for better visual appearance
#wedgeprops=dict(width=0.4): Creates a donut-style pie chart by specifying a width of 0.4.

# Add a legend with percentages
# Adjusts the bounding box of the legend to the right of the pie chart.
ax.legend(wedges, type_with_room, title="Rooms", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Reserved Rooms', fontsize=14)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A doughnut chart would better represent the type of rooms reserved.




##### 2. What is/are the insight(s) found from the chart?

Over half of the customers prefer room type A(64.71%) then D(19.91%).


##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

#### Chart - 14 - Correlation Heatmap

````python
hotel_data = df[['hotel','is_canceled','arrival_date_month','meal',
                                     'country','market_segment','distribution_channel',
                                     'is_repeated_guest', 'reserved_room_type',
                                     'assigned_room_type','deposit_type','agent',
                                     'customer_type','reservation_status',
                                     'stays_in_weekend_nights','stays_in_week_nights']]

# This class is used to convert categorical text data into numerical data for ML algorithms
labelencoder = LabelEncoder()
# Fits the label encoder to the unique values in the column and transforms the column into numeric labels.
hotel_data['hotel'] = labelencoder.fit_transform(hotel_data['hotel'])
hotel_data['arrival_date_month'] = labelencoder.fit_transform(hotel_data['arrival_date_month'])
hotel_data['meal'] = labelencoder.fit_transform(hotel_data['meal'])
hotel_data['country'] = labelencoder.fit_transform(hotel_data['country'])
hotel_data['market_segment']= labelencoder.fit_transform(hotel_data['market_segment'])
hotel_data['distribution_channel']=labelencoder.fit_transform(hotel_data['distribution_channel'])
hotel_data['is_repeated_guest'] = labelencoder.fit_transform(hotel_data['is_repeated_guest'])
hotel_data['reserved_room_type'] = labelencoder.fit_transform(hotel_data['reserved_room_type'])
hotel_data['assigned_room_type'] = labelencoder.fit_transform(hotel_data['assigned_room_type'])
hotel_data['deposit_type'] = labelencoder.fit_transform(hotel_data['deposit_type'])
hotel_data['agent'] = labelencoder.fit_transform(hotel_data['agent'])
hotel_data['customer_type'] = labelencoder.fit_transform(hotel_data['customer_type'])
hotel_data['reservation_status'] = labelencoder.fit_transform(hotel_data['reservation_status'])
hotel_data['stays_in_weekend_nights'] = labelencoder.fit_transform(hotel_data['stays_in_weekend_nights'])
hotel_data['stays_in_week_nights'] = labelencoder.fit_transform(hotel_data['stays_in_week_nights'])

hotel_data_cat = hotel_data[['hotel','is_canceled','arrival_date_month','meal',
                                     'country','market_segment','distribution_channel',
                                     'is_repeated_guest', 'reserved_room_type',
                                     'assigned_room_type','deposit_type','agent',
                                     'customer_type','reservation_status',
                                     'stays_in_weekend_nights','stays_in_week_nights']]

````

````python
""" Correlation Heatmap visualization code
correlation matrix is used to show the relationships between different categorical variables in the hotel_data_cat DataFrame."""
plt.figure(figsize=(14,14))
"""Calculate the Spearman correlation matrix for the DataFrame hotel_data_cat.
Spearman correlation method is used because it measures the monotonic relationship between variables."""
corr_categorical=hotel_data_cat.corr(method='spearman')
# Create a boolean mask for the upper triangle of the correlation matrix.
# np.triu generates an upper triangular matrix of ones
mask_categorical = np.triu(np.ones_like(corr_categorical, dtype=bool))
sns.heatmap(corr_categorical, annot=True, fmt=".2f", cmap='BrBG', vmin=-1, vmax=1, center= 0,
            square=True, linewidths=2, cbar_kws={"shrink": .5}).set(ylim=(16, 0))
plt.title("Correlation Matrix Spearman Method- Categorical Data ",size=16, weight='bold')
````

##### 1. Why did you pick the specific chart?

Heatmaps are better suited for  categorical correlationship.

##### 2. What is/are the insight(s) found from the chart?

There are strong correlation(positive) between reserved_room_type and assigned_room_type, also between market_segement and distribution channel. There is strong negative correlatin between reservation_status and is_canceled which makes sense as cancellation changes the resevation status.
